```python
import pandas as pd
import numpy as np
import random
```

```python
dataset = pd.read_csv('census.csv')
```

```python
dataset.shape
dataset.head()
```

```python
df_amostra_aleatoria_simples = dataset.sample(n=100, random_state=1)
df_amostra_aleatoria_simples.shape
df_amostra_aleatoria_simples.head()
```

```python
def amostra_aleatoria (dataset, amostras):
  return dataset.sample(n = amostras, random_state = 1)

df_amostra_aleatoria_simples = amostra_aleatoria(dataset, 100)
df_amostra_aleatoria_simples.shape
df_amostra_aleatoria_simples.head()
```

Amostragem Sistematica

```python
np.arange(68, len(dataset), step=325)
```

```python
def amostragem_sistematica(dataset, amostras): # funçao para selecionar X amostras
  intervalo = len(dataset) // amostras # calcula o intervalo \ formula. população / qtd amostras desejada = inteiro da divisão
  print(f"intervalo {intervalo}")
  random.seed(1) # random.seed() 0 ou 1 para trazer o mesmo resultado sempre
  inicio = random.randint(0, intervalo) # seleciona um numero aleatorio entre 0 e o intervalo
  print(f"inicio {inicio}")
  indices = np.arange(inicio, len(dataset), step = intervalo) # np.arange serve como for para iterar e armazenar na variavel os indices resultantes em um array
  print(f" indices {indices}")
  amostra_sistematica = dataset.iloc[indices] # com o valor da varivel indice, uso iloc[] para selecionar apenas os indices da variavel indices na tabela mãe e armazeno em amostra_sistematica
  return amostra_sistematica # retorna o resultado de amostra_sistematica para a função

df_teste = amostragem_sistematica(dataset, 10) # chama a função, passa a tabela toda e quantas amostras deseja retornar e armazena em df_teste
df_teste
```
